```python
#Import Libraries:
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Create Simple Dataset:
hours = np.array([2, 4, 6, 8, 10, 12, 14]).reshape(-1,1)
marks = np.array([50, 60, 75, 80, 84, 91, 97])

# hours, marks
```

```python
#Train a Linear Regression Model:
model = LinearRegression()
model.fit(hours, marks)
```

The Regression Line Equation is: ***y = mx + c***

Now, we will check how the model actually predict the linear regression line.

```python
#View the values:
print("m: ", model.coef_[0])
print("c: ", model.intercept_)

#So, the equation we get: y = 3.786*hours + 46.429
```

```python
from sys import hash_info
#Visualize the Best Fit Line:
plt.scatter(hours, marks, color='blue', label='Actual Data')

predicted_marks = model.predict(hours)
plt.plot(hours, predicted_marks, color='red', label='Best Fit Line')

plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours Vs Marks using Linear Regression Model")

plt.legend()
plt.grid(True)
plt.show()
```

```python
#Add New Student Data:

new_hours = np.array([[6.5]])
prediction = model.predict(new_hours)
print("Predicted Marks: ", prediction[0])
```
